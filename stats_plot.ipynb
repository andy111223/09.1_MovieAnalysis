```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Example DataFrame
data = {
    'title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E'],
    'release_date': ['2015-02-09', '2018-05-22', '2018-11-12', '2021-12-15', '2021-05-10'],
    'revenue': [500000, 1200000, 800000, 2000000, 1800000],
    'budget': [100000, 300000, 250000, 600000, 500000]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Step 1: Convert 'release_date' to datetime type
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Step 2: Extract the year from 'release_date'
df['year'] = df['release_date'].dt.year
```

```python
# Step 3: Group by 'year' and calculate average revenue and average budget
average_stats_per_year = df.groupby('year').agg(
    avg_revenue=('revenue', 'mean'),
    avg_budget=('budget', 'mean')
).reset_index()
```

```python
# Step 4: Plot the data
fig, ax = plt.subplots()
```

```python
# Plot the bar chart for average revenue
ax.bar(average_stats_per_year['year'], average_stats_per_year['avg_revenue'], color='lightblue', label='Average Revenue')
```

```python
# Plot the line chart for average budget
ax.plot(average_stats_per_year['year'], average_stats_per_year['avg_budget'], color='orange', marker='o', label='Average Budget')
```

```python
# Adding labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Amount (in USD)')
ax.set_title('Average Revenue and Budget per Year')
ax.legend()
```

```python
# Format the y-axis as currency
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '${:,.0f}'.format(x)))
```

```python
# Display the plot
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(average_stats_per_year['year'], average_stats_per_year['avg_revenue'], color='lightblue', label='Average Revenue')
ax.plot(average_stats_per_year['year'], average_stats_per_year['avg_budget'], color='orange', marker='o', label='Average Budget')
ax.set_xlabel('Year')
ax.set_ylabel('Amount (in USD)')
ax.set_title('Average Revenue and Budget per Year')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '${:,.0f}'.format(x)))
plt.show()
```
